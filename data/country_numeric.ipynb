```python
import pandas as pd
```

```python
df = pd.read_csv("https://gist.githubusercontent.com/tadast/8827699/raw/61b2107766d6fd51e2bd02d9f78f6be081340efc/countries_codes_and_coordinates.csv")
df.head()
```

```python
df_new = df[['Numeric code', 'Country']]

df_new['Numeric code'] = df_new['Numeric code'].str.strip().str.replace('"', '')
df_new['Country'] = df_new['Country'].str.strip().str.replace('"', '')

df_new['Numeric code'] = df_new['Numeric code'].astype(str).str.zfill(3)

country_dict = dict(zip(df_new['Numeric code'], df_new['Country']))

print(country_dict['004'])
```

```python
import json
with open("country_numeric.json", "w") as json_file:
    json.dump(country_dict, json_file, indent=4)
```
