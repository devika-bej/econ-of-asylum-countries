```python

import json

with open('conversion_numeric.json', 'r') as f:
    conversion_numeric = json.load(f)

inverted_conversion_numeric = {v: k for k, v in conversion_numeric.items()}
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/API_SP.POP.TOTL_DS2_en_csv_v2_84031.csv', index_col=0)
df.head()
```

```python
country_code_to_values = {}
for index, row in df.iterrows():
    country_code = row['Country Code']
    year_range = [str(year) for year in range(2000, 2021)]
    values = [row[year] for year in year_range]
    country_code_to_values[country_code] = values
```

```python
country_code_to_values['ABW']
```

```python
population_dataset = {}
for country_code, values in country_code_to_values.items():
    new_key = inverted_conversion_numeric.get(country_code, None)
    if new_key is not None:
        population_dataset[new_key] = values
    else:
        print(f"Key for country code '{country_code}' not found in inverted_conversion_numeric.")
```

```python

with open('population_dataset.json', 'w') as f:
    json.dump(population_dataset, f)
```
