```python
df = df_cleaned
df['Year'] = df['Year'].astype(str)

data_dict = {}

for index, row in df.iterrows():
    year = row['Year']
    country = row['origin']
    row_data = row.drop(['Year', 'origin']).to_dict()

    if year not in data_dict:
        data_dict[year] = {}

    data_dict[year][country] = row_data

```

```python
country_code_df = pd.read_csv("all.csv", usecols=['name', 'alpha-3', 'country-code'])
country_code_df.head()
```

```python
country_code_df.info()
```

```python
country_code_df['country-code'] = country_code_df['country-code'].astype(str).str.zfill(3)

country_code_dict = dict(zip(country_code_df['alpha-3'], country_code_df['country-code']))

print(country_code_dict)
```

```python
df_cleaned.head()
```

```python
df_cleaned['origin-code'] = df_cleaned['OriginISO'].map(country_code_dict)

df_cleaned['asylum-code'] = df_cleaned['AsylumISO'].map(country_code_dict)

df_cleaned.head()
```

```python
df_cleaned.isna().sum()
```

```python
df_cleaned[df_cleaned.isna().any(axis=1)]
```

```python
df_test = df_cleaned.dropna()
df_test.head()
```

```python
df_test.head()
```

```python
df_test.info()
```

```python
sorted_data = {}

for index, row in df_test.iterrows():
    year = row['Year']
    origin_code = row['origin-code']

    if year not in sorted_data:
        sorted_data[year] = {}

    sorted_data[year][origin_code] = {
        'origin': row['origin'],
        'OriginISO': row['OriginISO'],
        'OriginName': row['OriginName'],
        'asylum': row['asylum'],
        'AsylumISO': row['AsylumISO'],
        'AsylumName': row['AsylumName'],
        'AsylumRegion': row['AsylumRegion'],
        'PT': row['PT'],
        'Count': row['Count'],
        'asylum-code': row['asylum-code']
    }
```

```python
sorted_data.keys()
```

```python
sorted_data[2000]['004'] # country code for afghanistan
```

```python
# Above indicates that in the year 2000, 5 people left from afghanistan to zimbabwe
```

```python
incoming_data = {}

for index, row in df_test.iterrows():
    year = row['Year']
    asylum_code = row['asylum-code']

    if year not in incoming_data:
        incoming_data[year] = {}

    if asylum_code not in incoming_data[year]:
        incoming_data[year][asylum_code] = 0

    incoming_data[year][asylum_code] += row['Count']

# Now 'sorted_data' contains the dictionary with keys as years, and for each year,
# a dictionary with keys as 'asylum-code' and values as the count of people seeking asylum
```

```python
incoming_data[2000]['008']
```

```python
# incoming_data[2000]['010'] gives key error meaning that people seeking refuge
# in the country 010 (antartica) in year 2000 is 0
```

```python
import json

with open('outgoing_refugee_data.json', 'w') as json_file:
    json.dump(sorted_data, json_file)

with open('incoming_refugee_data.json', 'w') as json_file:
    json.dump(incoming_data, json_file)
```
