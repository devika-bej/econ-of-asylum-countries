```python
desired_subject_codes = ['NGDP', 'LUR', 'PCPIPCH', 'GGXWDN_NGDP', 'PPPPC', 'PPPGDP']

filtered_weo_data = weo_data[weo_data['WEO Subject Code'].isin(desired_subject_codes)]

filtered_weo_data.head()
```

```python
country_code_df.head()
```

```python
country_code_df['iso3'] = country_code_df['country-code'].astype(str)

country_code_df['iso3'] = country_code_df['iso3'].str.zfill(3)

country_code_df.head()
```

```python
alpha3_iso_dict = dict(zip(country_code_df['alpha-3'], country_code_df['iso3']))
# alpha3_iso_dict
```

```python
filtered_weo_data['iso3'] = filtered_weo_data['ISO'].map(alpha3_iso_dict)
filtered_weo_data.head()
```

```python
filtered_weo_data['iso3'].unique()
filtered_weo_data = filtered_weo_data.dropna(subset=['iso3'])
filtered_weo_data['iso3'].unique()
```

```python
subject_code_dfs = {}

for subject_code in desired_subject_codes:
    subject_df = filtered_weo_data[filtered_weo_data['WEO Subject Code'] == subject_code]
    subject_code_dfs[subject_code] = subject_df

subject_code_dfs.keys()
```

```python
subject_code_dfs['NGDP'].head()
```

```python
years = [str(year) for year in range(2000, 2024)]
print(years)
```

```python
for subject_code, df in subject_code_dfs.items():
    columns_to_keep = ['iso3'] + years

    subject_code_dfs[subject_code] = df[columns_to_keep]

subject_code_dfs.keys()
```

```python
subject_code_dfs['NGDP'].head()
```

```python
subject_code_results = {}
for subject_code, df in subject_code_dfs.items():
    subject_code_result = {}

    for year in range(2000, 2024):
        year_df = df[df[str(year)].notna()]
        iso3_dict = year_df.set_index('iso3')[str(year)].to_dict()
        subject_code_result[str(year)] = iso3_dict
    subject_code_results[subject_code] = subject_code_result

subject_code_results.keys()
```

```python
len(subject_code_results['NGDP']['2000'].keys())
```

```python
subject_code_results = {}

for subject_code, df in subject_code_dfs.items():
    subject_code_result = {}

    for iso3 in df['iso3'].unique():
        iso3_result = {}

        for year in range(2000, 2024):
            year_iso3_df = df[(df['iso3'] == iso3) & (df[str(year)].notna())]
            year_iso3_dict = year_iso3_df.set_index('iso3')[str(year)].to_dict()
            iso3_result[str(year)] = year_iso3_dict

        subject_code_result[iso3] = iso3_result

    subject_code_results[subject_code] = subject_code_result

subject_code_results.keys()
```

```python
for subject_code, result in subject_code_results.items():
    for iso3, year_dict in result.items():
        flattened_year_dict = {}
        for year, value_dict in year_dict.items():
            if value_dict:
                flattened_year_dict[year] = next(iter(value_dict.values()))
            else:
                flattened_year_dict[year] = None
        subject_code_results[subject_code][iso3] = flattened_year_dict


subject_code_results.keys()
```

```python
subject_code_results['LUR']['012']
```

```python
years_not_none = {}

for subject_code, result in subject_code_results.items():
    years_not_none[subject_code] = {}

    for iso3, year_dict in result.items():
        # years_not_none[subject_code][iso3] =
        subject_code_results[subject_code][iso3]["present"] = [year for year, value in year_dict.items() if value is not None]
        subject_code_results[subject_code][iso3]["present"].remove("present")

(subject_code_results['NGDP']['004'])
```

```python
import json

for subject_code, data in subject_code_results.items():
    filename = f"{subject_code}.json"
    with open(filename, "w") as json_file:
        json.dump(data, json_file)

```
