# Incoming Refugee Data

```python
import json
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

with open('/content/incoming_refugee_data.json', 'r') as file:
    data = json.load(file)

all_values = []
for year_data in data.values():
    all_values.extend(year_data.values())

hist, bins, _ = plt.hist(all_values, bins=50, color='skyblue', edgecolor='black')
plt.title('Histogram of Incoming Refugees values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.ylim(0, max(hist) * 1.1)
plt.yticks(range(0, int(max(hist) * 1.1), int(max(hist) / 10)))

plt.tight_layout()
plt.show()
```

# Outgoing Refugee Data


```python

with open('/content/outgoing_refugee_data.json', 'r') as file:
    data = json.load(file)

all_values = []
for year_data in data.values():
    for country_code, country_data in year_data.items():
        for value_data in country_data.values():
          all_values.append(value_data)

plt.hist(all_values, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
filtered_values = [value for value in all_values if 0 <= value <= 20000]

plt.hist(filtered_values, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Values (0 to 20,000)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
filtered_values = [value for value in all_values if 0 <= value <= 500]

plt.hist(filtered_values, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Values (0 to 500)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
